```
%matplotlib inline

%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
%config InlineBackend.rc = {'font.size': 10, 'figure.figsize': (10.0, 6.0), 'figure.facecolor': (1, 1, 1, 0), 'figure.subplot.bottom': 0.125, 'figure.edgecolor': (1, 1, 1, 0), 'figure.dpi': 72}
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
df1=pd.read_csv("sample_data/Mamooth.csv")#specify the path of csv file
df2=pd.read_csv("sample_data/Apollo Medical.csv")

```

```
df1
df2
```

```
df3 = df1.set_index("Unnamed: 0", drop = False)
df4=df2.set_index("Unnamed: 0", drop = False)
df4
```

```
df3.drop(["Unnamed: 0"],axis=1)
df4.drop(["Unnamed: 0"],axis=1)
df4.loc["Total current assets",:]
```

```
k1=list(df3.loc["Total current assets",:])
k2=list(df4.loc["Total current assets",:])
a1=list(df3.loc["Cash and short-term investments",:])
a2=list(df4.loc["Cash and short-term investments",:])
b1=list(df3.loc["Total debt",:])
b2=list(df4.loc["Total debt",:])
c1=list(df3.loc["Total liabilities",:])
c2=list(df4.loc["Total liabilities",:])
d1=list(df3.loc["Shareholders Equity",:])
d2=list(df4.loc["Shareholders Equity",:])
```

```
l1=list(df3.columns.values)
l1
```

```
plt.plot(l1[1:], k1[1:], '--')
plt.plot(l1[1:], k2[1:len(l1)], '--')
plt.axis([l1[1], l1[-1],min(min(k1[1:]),min(k2[1:len(l1)])),max(max(k1[1:]),max(k2[1:len(l1)]))+2*min(k1[1:])])
plt.legend(["Mamooth Industries","Apollo Medical"])

plt.title("Comparison")
plt.xlabel("Quarterly")
plt.ylabel("Total current assets")
plt.show()
```

```
plt.plot(l1[1:], a1[1:], '--')
plt.plot(l1[1:], a2[1:len(l1)], '--')
plt.axis([l1[1], l1[-1],min(min(a1[1:]),min(a2[1:len(l1)])),max(max(a1[1:]),max(a2[1:len(l1)]))+2*min(a1[1:])])
plt.legend(["Mamooth Industries","Apollo Medical"])

plt.title("Comparison")
plt.xlabel("Quarterly")
plt.ylabel("Cash and short-term investments")
plt.show()
```

```
plt.plot(l1[1:], b1[1:], '--')
plt.plot(l1[1:], b2[1:len(l1)], '--')
plt.axis([l1[1], l1[-1],min(min(b1[1:]),min(b2[1:len(l1)])),max(max(b1[1:]),max(b2[1:len(l1)]))+2*min(b1[1:])])
plt.legend(["Mamooth Industries","Apollo Medical"])

plt.title("Comparison")
plt.xlabel("Quarterly")
plt.ylabel("Total debt")
plt.show()
```

```
plt.plot(l1[1:], c1[1:], '--')
plt.plot(l1[1:], c2[1:len(l1)], '--')
plt.axis([l1[1], l1[-1],min(min(c1[1:]),min(c2[1:len(l1)])),max(max(c1[1:]),max(c2[1:len(l1)]))+2*min(c1[1:])])
plt.legend(["Mamooth Industries","Apollo Medical"])

plt.title("Comparison")
plt.xlabel("Quarterly")
plt.ylabel("Total liabilities")
plt.show()
```

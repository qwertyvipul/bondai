```
df1 = pd.read_csv('sample_data/malibu.csv')#getting csv file
```

```

df2 = df1.set_index("Unnamed: 0", drop = True)#setting index
```

```
df2
```

```
#calculating debt ratio (debt ratio is used to find the stability of company)
#debt ratio=Long-term liabilities/Total assets
debt_ratio=[]
l1=list(df2.columns.values)#getting list of column headings

long_term_liabilities=list(df2.loc["Total non-current liabilities",:])
Total_assets=list(df2.loc["Total assets",:])

count=len(l1)
for i in range(count):
  debt_ratio.append(long_term_liabilities[i]/Total_assets[i])
debt_ratio #Higher the ratio higher the risk of company going to default 
  



```

```

plt.plot(l1[:10], debt_ratio[:10], '--')
plt.axis([l1[0], l1[9],int(min(debt_ratio[:10])),int(max(debt_ratio[:10])+max(debt_ratio[:10]))])


plt.title("Malibu Boats Financial Stability Analysis")
plt.xlabel("Quarterly")
plt.ylabel("Debt Ratio(Increasing risk of default)")
plt.show()
```
